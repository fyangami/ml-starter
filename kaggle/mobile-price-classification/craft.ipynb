```python
# 数据预处理
train_data = train_data.to_numpy()
train_label = train_data[:, -1:]
train_data = train_data[:, :-1]
test_data = test_data.to_numpy()
test_label = test_data[:, -1:]
test_data = test_data[:, :-1]
```

```python
import sys
sys.path.append('/home/fyang/code/ml/ml-starter')

from models.logistic_regression import LogisticClassifier
from utils.preprocess import std

# 先做1-1回归
train_label = (train_label == 1).astype('uint8')
test_label = (test_label == 1).astype('uint8')

train_data = std(train_data)
test_data = std(test_data)

model = LogisticClassifier(eta=1e-1, epochs=100, batch_size=200)
model.fit(train_data, train_label, test_data, test_label)
from matplotlib import pyplot as plt
plt.plot(model.losses)
model2 = LogisticClassifier(eta=1e-1, epochs=100, batch_size=30)
model2.fit(train_data, train_label, test_data, test_label)
plt.show()
plt.plot(model2.losses)
```
